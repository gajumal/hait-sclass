```python
#多項式回帰(2次)

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

def quad_pca(number, xlim_l, xlim_r):          #引数1:pca_number,2:グラフの左端,3:グラフの右端
    pca_x = data0_pca[:,[number]]          #data取り出し
    quad = PolynomialFeatures(degree=2)    #変数変換
    quad_pcax = quad.fit_transform(pca_x) 

    #モデルを学習
    quad_pcax_train, quad_pcax_test, df_target_train, df_target_test = train_test_split(quad_pcax,
                                                       df_target, 
                                                       test_size= 0.3,
                                                        random_state = 1)
                                                       
    lr_x = LinearRegression()                      #線形回帰
    lr_x.fit(quad_pcax_train, df_target_train)     #モデル学習
    
    print(lr_x.score(quad_pcax_train, df_target_train))  #決定係数を出力
    print(lr_x.score(quad_pcax_test, df_target_test))
    
    #散布図に回帰曲線
    plt.scatter(pca_x, df_target, s=10, color='lightgray')
    x = np.arange(-10,10,1)[:, np.newaxis]        #xを生成
    x_quad = quad.fit_transform(x)                      #xを2次に変換
    plt.plot(x, lr_x.predict(x_quad), color='red')
    plt.xlim([xlim_l,xlim_r])
    plt.show()
```

```python
#多項式回帰(3次)

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

def cubic_pca(number, xlim_l, xlim_r):          #引数1:pca_number,2:グラフの左端,3:グラフの右端
    pca_x = data0_pca[:,[number]]              #data取り出し
    cubic = PolynomialFeatures(degree=3)        #変数変換
    cubic_pcax = cubic.fit_transform(pca_x) 

    #モデルを学習
    cubic_pcax_train, cubic_pcax_test, df_target_train, df_target_test = train_test_split(cubic_pcax,
                                                       df_target, 
                                                       test_size= 0.3,
                                                        random_state = 1)
                                                       
    lr_x = LinearRegression()                      #線形回帰
    lr_x.fit(cubic_pcax_train, df_target_train)     #モデル学習
    
    print(lr_x.score(cubic_pcax_train, df_target_train))  #決定係数を出力
    print(lr_x.score(cubic_pcax_test, df_target_test))
    
    #散布図に回帰曲線
    plt.scatter(pca_x, df_target, s=10, color='lightgray') #散布図をプロット
    x = np.arange(-10,10,1)[:, np.newaxis]              #xを生成
    x_cubic = cubic.fit_transform(x)                      #xを3次に変換
    plt.plot(x, lr_x.predict(x_cubic), color='red')
    plt.xlim([xlim_l,xlim_r])
    plt.show()
```

```python
#多項式回帰、正則化

#特徴量抽出
pca0 = df_data_pca.loc[:,['pca_0']].values
pca3 = df_data_pca.loc[:,['pca_3']].values
pca9 = df_data_pca.loc[:,['pca_9']].values

#変数変換
from sklearn.preprocessing import PolynomialFeatures
cubic = PolynomialFeatures(degree=3)

pca0_cubic = cubic.fit_transform(pca0)
pca3_cubic = cubic.fit_transform(pca3)
pca9_cubic = cubic.fit_transform(pca9)

#変数を連結
pca_sel = np.hstack((pca0_cubic, pca3_cubic, pca9_cubic))

#dataを分割
from sklearn.model_selection import train_test_split
pca_sel_train, pca_sel_test, df_target_train, df_target_test = train_test_split(pca_sel,
                                                                                df_target,
                                                                                test_size=0.3
                                                                                )

#正則化ridge,lasso
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
lr = LinearRegression()
ridge = Ridge(alpha=100)
lasso = Lasso(alpha=300)
lr.fit(pca_sel_train, df_target_train)
ridge.fit(pca_sel_train, df_target_train)
lasso.fit(pca_sel_train, df_target_train)
```

```python
#決定係数を出力

print(lr.score(pca_sel_train, df_target_train))
print(lr.score(pca_sel_test, df_target_test))
print(ridge.score(pca_sel_train, df_target_train))    #ridge
print(ridge.score(pca_sel_test, df_target_test))
print(lasso.score(pca_sel_train, df_target_train))    #lasso
print(lasso.score(pca_sel_test, df_target_test))
```

```python
#交差検証法

#data分割方法を指定
from sklearn.model_selection import KFold
kf = KFold(n_splits=5, shuffle=True, random_state=0)

#交差検証を実行
from sklearn.model_selection import cross_val_score
score_lr = cross_val_score(lr, pca_sel, df_target, cv=kf)
score_ridge = cross_val_score(ridge, pca_sel, df_target, cv=kf)
score_lasso = cross_val_score(lasso, pca_sel, df_target, cv=kf)
```

```python
#交差検証精度の平均を出力
#交差検証精度の標準偏差を出力

print(score_lr.mean())
print(score_lr.std())
print(score_ridge.mean())    #ridge
print(score_ridge.std())
print(score_lasso.mean())    #lasso
print(score_lasso.std())
```
